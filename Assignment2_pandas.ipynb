# City-Wise Analysis
Find total sales for each city in each month using groupby()

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# Group by city and sum monthly sales
city_monthly_sales = store_sales.groupby('city')[months].sum()
city_monthly_sales
```

Find the average monthly sales per city

```python
average_monthly_sales = store_sales.groupby('city')[months].mean()
average_monthly_sales
```

# Aggregation and Analysis
Using groupby("city"), compute:
Mean, Max, and Min for each month.

```python
monthly_stats = store_sales.groupby('city')[['Jan','Feb','Mar','Apr']].agg(['mean','max','min'])
print(monthly_stats)
```

Find the city with the highest average March sales

```python
march_avg = store_sales.groupby('city')['Mar'].mean()
top_city_march = march_avg.idxmax()
top_value = march_avg.max()
print(f"City with highest average March sales: {top_city_march} ({top_value})")
```

Calculate the overall average sales per city (across all months)

```python
overall_avg_sales = store_sales.groupby('city')[['Jan','Feb','Mar','Apr']].mean().mean(axis=1)
print(overall_avg_sales)
```

Identify the month with the highest overall sales

```python
monthly_totals = store_sales[['Jan','Feb','Mar','Apr']].sum()
top_month = monthly_totals.idxmax()
top_value = monthly_totals.max()

print(f"Month with highest overall sales: {top_month} ({top_value})")
```

For each city, find the month with maximum total sales.

```python
city_monthly_totals = store_sales.groupby('city')[['Jan','Feb','Mar','Apr']].sum()
max_month_per_city = city_monthly_totals.idxmax(axis=1)
print(max_month_per_city)
```
